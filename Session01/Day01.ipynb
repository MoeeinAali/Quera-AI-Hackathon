```python
a = np.random.rand(1000)
b = np.random.rand(1000) + 1.1
x = np.concatenate((a, a, b, b))
y = np.concatenate((a, b, a, b))
df = pd.DataFrame({'x': x, 'y': y})
plt.scatter(x, y)
plt.show()
```

```python
k = KMeans(n_clusters=2)
k.fit(df)
plt.scatter(df.x, df.y, c=k.labels_)
plt.show()
```

```python
k = KMeans(n_clusters=3)
k.fit(df)
plt.scatter(df.x, df.y, c=k.labels_)
```

```python
k = KMeans(n_clusters=4)
k.fit(df)
plt.scatter(df.x, df.y, c=k.labels_)
```

```python
elbow = []
for i in range(1, 11):
    k = KMeans(n_clusters=i)
    k.fit(df)
    elbow.append(k.inertia_)

plt.plot(range(1, 11), elbow)
```

```python
# without Scaling

a = [1, 1, 1, 1, 1, 1, 2, 2, 2, 2]
b = [
    110000, 111000, 112000, 200000, 202000, 198000, 110000, 111000, 112000,
    115500
]
features = pd.DataFrame({'x': a, 'y': b})
plt.scatter(features.x, features.y)
plt.xlim(0, 220000)
plt.ylim(0, 220000)
plt.show()
```

```python
# With Scaling - without library 

a = [1, 1, 1, 1, 1, 1, 2, 2, 2, 2]
b = [
    110000, 111000, 112000, 200000, 202000, 198000, 110000, 111000, 112000,
    115500
]
features = pd.DataFrame({'x': a, 'y': b})
features.x = (features.x - features.x.min())/(features.x.max() - features.x.min())
features.y = (features.y - features.y.min())/(features.y.max() - features.y.min())
plt.scatter(features.x, features.y)
plt.show()
```

```python
# With Scaling - with library 

from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
a = [1, 1, 1, 1, 1, 1, 2, 2, 2, 2]
b = [
    110000, 111000, 112000, 200000, 202000, 198000, 110000, 111000, 112000,
    115500
]
features = pd.DataFrame({'x': a, 'y': b})
features = pd.DataFrame(sc.fit_transform(features),columns=features.columns)
plt.scatter(features.x,features.y)
plt.show()
```
